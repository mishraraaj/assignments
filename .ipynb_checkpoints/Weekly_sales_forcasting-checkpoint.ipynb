```python
def trend_for_dept(df,store_dept):
    """Weekly Trend For one Department"""
    try:
        plt.figure(figsize=(12, 4))
        plt.plot(df[df['Store_dept']==store_dept]['Weekly_Sales'])
        plt.title('Sales (Weekly data)')
        plt.ylabel("Weekly Sales")
        plt.xlabel("Dates")
        plt.grid(True)
        plt.show()
    except Exception as e:
        print(e)
```

```python
from sklearn.metrics import r2_score, median_absolute_error, mean_absolute_error
from sklearn.metrics import median_absolute_error, mean_squared_error, mean_squared_log_error

def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
def moving_average(series, n):
    """
        Calculate average of last n observations
    """
    return np.average(series[-n:])
```

```python
def plotMovingAverage(series, window, plot_intervals=False, scale=1.96, plot_anomalies=False):

    """
        series - dataframe with timeseries
        window - rolling window size 
        plot_intervals - show confidence intervals
        plot_anomalies - show anomalies 

    """
    rolling_mean = series.rolling(window=window).mean()

    plt.figure(figsize=(15,5))
    plt.title("Moving average\n window size = {}".format(window))
    plt.plot(rolling_mean, "g", label="Rolling mean trend")

    # Plot confidence intervals for smoothed values
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bond = rolling_mean - (mae + scale * deviation)
        upper_bond = rolling_mean + (mae + scale * deviation)
        plt.plot(upper_bond, "r--", label="Upper Bond / Lower Bond")
        plt.plot(lower_bond, "r--")
        
        # Having the intervals, find abnormal values
        if plot_anomalies:
            anomalies = pd.DataFrame(index=series.index, columns=series.columns)
            anomalies[series<lower_bond] = series[series<lower_bond]
            anomalies[series>upper_bond] = series[series>upper_bond]
            plt.plot(anomalies, "ro", markersize=10)
        
    plt.plot(series[window:], label="Actual values")
    plt.legend(loc="upper left")
    plt.grid(True)
```

```python

def scatterplot(dataset,numerical_columns,dependent_variable):
    for col in numerical_columns:
        if col != dependent_variable:
            fig = plt.figure(figsize=(9, 6))
            ax = fig.gca()
            feature = dataset[col]
            label = dataset[dependent_variable]
            correlation = feature.corr(label)
            plt.scatter(x=feature, y=label)
            plt.xlabel(col)
            plt.ylabel(dependent_variable)
            ax.set_title('price vs ' + col + '- correlation: ' + str(correlation))
            z = np.polyfit(dataset[col], dataset[dependent_variable], 1)
            y_hat = np.poly1d(z)(dataset[col])

            plt.plot(dataset[col], y_hat, "r--", lw=1)

    plt.show()
```

# Feature Engineering

```python
previous_year_data=merge_data_train
```

```python
def weighted_last_years_sales(df,curr_year,alpha=0.5):
    
    store_wise_sale_prev1 = df[df.Year==curr_year-1].groupby(df['Store_dept'])['Weekly_Sales'].mean()
    store_wise_sale_prev2 = df[df.Year==curr_year-2].groupby(df['Store_dept'])['Weekly_Sales'].mean()

    Last_year_weighted_avg_sales=(alpha*store_wise_sale_prev1) + ((1-alpha) * store_wise_sale_prev2)
    return Last_year_weighted_avg_sales.reset_index().rename(columns={"Weekly_Sales":"Last_year_sales"})
```

```python
Last_year_weighted_avg_sales=weighted_last_years_sales(merge_data_train,2012,0.7)

merge_data_train=pd.merge(merge_data_train,Last_year_weighted_avg_sales,on=['Store_dept'],how='left')
```

```python
Last_year_weighted_avg_sales=weighted_last_years_sales(previous_year_data,2013,0.7)

merge_data_test=pd.merge(previous_year_data,Last_year_weighted_avg_sales,on=['Store_dept'],how='left').drop(columns='Weekly_Sales')
```

# Imputation by grouping on the bases of day and store No and then interpolating values  
Since it has maximum correlation with day and store

```python
def simple_impute(df,cols=['Day','Store','MarkDown1','MarkDown5']):
    df[['Day','Store','MarkDown1','MarkDown5']]= df[['Day','Store','MarkDown1','MarkDown5']].groupby(['Day']).apply(lambda group: group.interpolate(method='index').ffill().bfill())
    imputer = KNNImputer(n_neighbors=3, weights='uniform', metric='nan_euclidean')
    X=df[df.describe().columns.to_list()].values
    imputer.fit(X)
    Xtrans = imputer.transform(X)
    return Xtrans
```

```python
Xtrans=simple_impute(merge_data_train)
merge_data_train[merge_data_train.describe().columns.to_list()]=pd.DataFrame(Xtrans,columns=merge_data_train.describe().columns.to_list())
```

```python
Xtrans=simple_impute(merge_data_test)
merge_data_test[merge_data_test.describe().columns.to_list()]=pd.DataFrame(Xtrans,columns=merge_data_test.describe().columns.to_list())
```
